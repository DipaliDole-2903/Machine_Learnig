# Estimate weight of car

# step 1 - Data ingestion

```python
import pandas as pd
df = pd.read_csv("Cars93.csv", na_values = ["","NA"], keep_default_na=False)
df.head()
```

# step 2 - Perform basic data quality checks 

```python
df.shape
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates(keep="first").reset_index(drop=True)
df.shape
```

```python
m = df.isna().sum()
m[m>0]
```

```python
m[m>0] / len(df)
```

```python
df.info
```

```python
c= df.select_dtypes(include="object").nunique()
c
```

```python
cardinality = c / len(df)
cardinality
```

```python
high_card_cols = cardinality[cardinality >= 0.9].index
high_card_cols
```

```python
df = df.drop(columns = high_card_cols)
df.head()
```

# Step 3 - Seperate X and Y (Weight)
id dropped because it is a serial no Statistically insignificant

```python
X = df.drop(columns=["id", "Weight"])
Y = df["Weight"]
```

```python
X.head()
```

```python
Y.head()
```

# Step 4 - Apply Train Test Split

```python
from sklearn.model_selection import train_test_split

xtrain, xtest, ytrain, ytest = train_test_split(X, Y, test_size=0.2, random_state= 21 )
```

```python
xtrain.shape
```

```python
xtest.shape
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
xtest.head()
```

```python
ytest.head()
```

# Step 5- Build Preprocessing pipeline for X

```python
from sklearn.pipeline import make_pipeline # To combine preprocessing steps 
from sklearn.impute import SimpleImputer # To replace missing values
from sklearn.preprocessing import StandardScaler, OneHotEncoder # Preprocessing 
from sklearn.compose import ColumnTransformer # Combine num_pipe and Cat_pipe
```

```python
num_pipe = make_pipeline(
    SimpleImputer(strategy= "mean"),
    StandardScaler()
)
```

```python
cat_pipe = make_pipeline(
    SimpleImputer(strategy="most_frequent"),
    OneHotEncoder(handle_unknown="ignore", sparse_output=False, drop="first")
)
```

```python
cat_cols = xtrain.select_dtypes(include = "object").columns.tolist()
cat_cols
```

```python
num_cols = xtrain.select_dtypes(include="number").columns.tolist()
num_cols
```

```python
pre = ColumnTransformer(
    [
        ("num", num_pipe,num_cols),
        ("cat", cat_pipe, cat_cols )
    ]
).set_output(transform="pandas")

pre.fit(xtrain)
```

```python
xtrain_pre = pre.transform(xtrain)
xtrain_pre.head()
```

```python
xtest_pre = pre.transform(xtest)
xtest_pre.head()
```

# preprocessing completed

# Step 6 Build the model

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(xtrain_pre, ytrain)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
coef = pd.Series(model.coef_, index=xtrain_pre.columns)
coef
```

# step 7 - Model Evaluation

```python
# R2 score on train
model.score(xtrain_pre, ytrain)
```

```python
# R2 score on test 
model.score(xtest_pre, ytest)
```

```python
from sklearn.metrics import (
    root_mean_squared_error,
    mean_absolute_error,
    mean_absolute_percentage_error,
    r2_score
)

def evaluate_model(model, x, y):
    ypred = model.predict(x)
    rmse = root_mean_squared_error(y, ypred)
    mae = mean_absolute_error(y, ypred)
    mape = mean_absolute_percentage_error(y, ypred)
    r2 = r2_score(y, ypred)
    print(f"RMSE : {rmse:.2f}")
    print(f"MAE : {mae:.2f}")
    print(f"MAPE : {mape:.2%}")
    print(f"R2 : {r2:.2%}")
```

```python
print("Train Results : ")
evaluate_model(model, xtrain_pre, ytrain)
```

```python
print("test Results : ")
evaluate_model(model, xtest_pre, ytest)
```
